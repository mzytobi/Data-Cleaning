```python
def check_outlier(data,var):
    q1,q3=np.percentile(data[var],[25,75])
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr
    return(lower,upper)
```

```python
check_outlier(data,"Price")
```

```python
check_outlier(data,"Landsize")
```

```python
def drop_outlier(data,var):
    q1,q3=np.percentile(data[var],[25,75])
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr
    
    data=data[data[var]<upper]
    data=data[data[var]>lower]
    data.reset_index(drop=True,inplace=True)
    return data
```

```python
data1=drop_outlier(data,"Price")
data1
```

```python
def cap_outlier(data,var):
    q1,q3=np.percentile(data[var],[25,75])
    iqr=q3-q1
    lower=q1-1.5*iqr
    upper=q3+1.5*iqr
    
    data[var].values[data[var].values>upper]=upper
    data[var].values[data[var].values<lower]=lower
    return data
```
